## Sine Waves

A sinusoidal wave or sine wave is characterized by three parameters:
<ul>
<li>Frequency (f, in Hertz)
<li>Phase ($\phi$ in radians)
<li>Amplitude (wathever the wave unit)
</ul>

<p>Mathematically:

<p>$$x(t) = A \sin(2 \, \pi \, f \, t + \phi)$$

<p>The sine wave is <em>periodic</em> and its <em>period</em> (in seconds) is:

<p>$$T = \frac{1}{f}$$

<p>Here is an example:


```python
f = 1
A = 1
phi = 0
t = np.linspace(0, 5, 500)              # the prefix np. invokes functions in the numpi library
plt.plot(t, A*np.sin(2*np.pi*f*t+phi))  # the prefix plt. invokes functions in the matplotlib library
plt.title('Sine Wave, f = %d Hz' % f);
plt.xlabel('t (s)');
plt.ylabel('x(t)');
plt.grid(1)
```

### Understanding Phase

The effect of phase can be understood in light of the trigonometric identity:

<p>$$\sin(a+b) = \sin a \cos b + \sin b \cos a$$

<p>If we let $a = 2 \, \pi \, f \, t$ and $b = \phi$ then:

<p>$$\sin(2 \, \pi \, f \, t + \phi) = \cos \phi \sin(2 \, \pi \, f \, t) + \sin \phi \cos(2 \, \pi \, f \, t)$$

## Summing Sine Waves with Same Frequencies

When sine waves <b>with same frequency</b> are added together the result is still periodic with same frequency. 

<p>If the <b>phase is also the same</b> then the it's easy:

<p>$$A \sin(2 \, \pi \, f \, t+ \phi) + B \sin(2 \, \pi \, f \, t+ \phi) = (A + B) \sin(2 \, \pi \, f \, t + \phi)$$

```python
f = 1
A = 1
B = 1
phi = 0
t = np.linspace(0, 5, 500)
plt.plot(t, A*np.sin(2*np.pi*f*t+phi) + B*np.sin(2*np.pi*f*t+phi))
plt.title('Sine Wave, f = %d Hz' % f);
plt.xlabel('t (s)');
plt.ylabel('x(t)');
plt.grid(1)
```

<p>If the <b>phase is different</b> but the <b>amplitude is the same</b>:

<p>$$\begin{align*}
\sin(2 \, \pi \, f \, t) + \sin(2 \, \pi \, f \, t + \phi) &= \sin(2 \, \pi \, f \, t) + \cos(\phi) \sin(2 \, \pi \, f \, t) + \sin(\phi) \cos(2 \, \pi \, f \, t) \\
&= (1 + \cos \phi) \sin(2 \, \pi \, f \, t) + \sin \phi \cos(2 \, \pi \, f \, t)
\end{align*}$$

```python
f = 1
phi = np.pi/2
t = np.linspace(0, 5, 500)
plt.plot(t, np.sin(2*np.pi*f*t) + np.sin(2*np.pi*f*t+phi))
plt.title('Sine Wave, f = %d Hz' % f);
plt.xlabel('t (s)');
plt.ylabel('x(t)');
plt.grid(1)
```
