```python
t = var('t') 
var('R, L, C, t0, q0, I0');q = function('q')(t) 
assume(R>0);assume(L>0);assume(C>0);assume(t>0);
assume(4*L>C*R^2) 
RLC =R*diff(q,t,1)+L*diff(q,t,2)+q/C == 0
LC=LC = L*diff(q,t,2)+q/C == 0
```

```python
rlc = desolve(RLC, q, ivar=t, ics=[t0,q0,I0])
RLC
```

```python
q_RLC = desolve(RLC, q, ivar=t, ics=[t0,q0,I0])
```

```python
I=diff(q_RLC, t, 1)
```

```python
plotq_RLC=q_RLC.substitute(t0=0, q0=1e-6, I0=0, L=12e-2, C=1.5e-6, R=40);
```

```python
plotq_RLC
```

```python
qRLC=plotq_RLC
show(plot(plotq_RLC, (0, 0.01)),frame=True,figsize=4,axes=True) 
```
